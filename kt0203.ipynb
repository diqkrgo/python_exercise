• Q) 리스트 원소의 최솟값 구하기 함수 min_of( )와 최댓값 구하기 함수 max_of()을 작성하라.
• t = (4, 7, 5.6, 2, 3.14, 1)
• s = 'string'
• a = ['DTS', 'AAC', 'FLAC']

```python
# 최댓값 찾기
def max_of(a):
    max = a[0]
    for i in range(1, len(a)):
        if a[i] > max:
            max = a[i]
    return max

t = (4, 7, 5.6, 2, 3.14, 1)
s = 'string'
a = ['DTS', 'AAC', 'FLAC']

print(f'{max_of(t)}')
print(f'{max_of(s)}')
print(f'{max_of(a)}')
```

```python
# 최솟값 찾기
def min_of(a):
    min = a[0]
    for i in range(1, len(a)):
        if a[i] < min:
            min = a[i]
    return min

t = (4, 7, 5.6, 2, 3.14, 1)
s = 'string'
a = ['DTS', 'AAC', 'FLAC']

print(f'{min_of(t)}')
print(f'{min_of(s)}')
print(f'{min_of(a)}')
```

```python
# 일단 역순 생성하는 함수 만들기
def reverse_list(a):
    n = len(a)
    for i in range(n//2):
        a[i],a[n-i-1] = a[n-i-1], a[i]
    
a = [2, 5, 1, 3, 9, 6, 7]
reverse_list(a)
print(a)  
```

```python
print('리스트를 역순으로 출력합니다. ')
num = int(input('원소수를 입력하세요. :'))
x = [None] * num

for i in range(num):
    x[i] = int(input(f'x[{i}]값을 입력하세요. :'))
    
reverse_list(x)
    
print(f'리스트를 역순으로 출력합니다.')
for i, value in enumerate(x):
    print(f'x[{i}] = {value}')
```

## 이진검색 binary search

```python
def bin_search(a, key):
    pl = 0 # 맨 왼쪽
    pr = len(a)-1 # 맨 오른쪽
    
    while True:
        pc = (pl+pr)//2 # pc(중앙값 지정)
        if a[pc] == key: # 만약 a리스트의 pc가 key라면? pc값을 리턴
            return pc
        elif a[pc] < key: # 만약 pc가 key보다 작다면 pl을 pc+로
            pl = pc+1
        else:    #  만약 pc가 key보다 크다면? pr을 pc-1로 
            pr = pc-1
        
        if pl > pr: # 그래서 pl이 pr보다 크다 == pl=pr=pc라면 -1을 리턴하고 while문 탈출!
            break
    return -1 
```

### 0부터 9까지의 숫자 중 일부가 들어있는 배열 numbers가 매개변수로 주어집니다. numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 return 하도록 solution 함수를 완성해주세요.

- 제한사항
- 1 ≤ numbers의 길이 ≤ 9
- 0 ≤ numbers의 모든 수 ≤ 9
- numbers의 모든 수는 서로 다릅니다.

### 입출력 예

|numbers|result
|-------|-------
|[1,2,3,4,6,7,8,0]|14
|[5,8,4,0,6,7,9]|6

```python
def solution(numbers):
    n = list(range(1,10))
    answer = set(n)-set(numbers)
    return sum(answer)

numbers = [5,8,4,0,6,7,9]
print(solution(numbers))
```
